```python
faqs="""Machine learning (ML) and deep learning (DL) are transformative branches of artificial intelligence (AI), each playing a critical role in enabling computers to learn from data and improve their performance without explicit programming.
Machine learning is a broad field that encompasses various algorithms designed to identify patterns, classify data, or predict outcomes based on historical information.
These algorithms are typically divided into supervised learning, where the model is trained on labeled data; unsupervised learning, where it finds hidden patterns in unlabeled data; and reinforcement learning, which uses a system of rewards and penalties to learn optimal behaviors.
Common applications of ML include fraud detection, recommendation systems, and predictive maintenance.
On the other hand, deep learning is a specialized subset of ML that focuses on artificial neural networks with multiple layers—commonly referred to as deep neural networks.
These networks are inspired by the human brain’s structure and function, allowing them to model complex, high-dimensional relationships in data.
Deep learning excels in processing unstructured data such as images, audio, and text, making it the driving force behind recent advancements in computer vision, natural language processing, and speech recognition.
Techniques like convolutional neural networks (CNNs) are pivotal in image recognition tasks, while recurrent neural networks (RNNs) and transformers are crucial for sequential data like language modeling.
Despite its success, deep learning often requires vast amounts of data and computational power, which can be a limiting factor.
Together, ML and DL are reshaping industries by enabling smarter systems, automating processes, and unlocking new possibilities in areas like healthcare, finance, transportation, and entertainment.
Their continued development promises even more sophisticated applications, heralding a future where AI seamlessly integrates into daily life."""
```

```python
print(type(faqs))
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
tokenizer=Tokenizer()
tokenizer.fit_on_texts([faqs])
tokenizer.word_index
```

```python
input_sequences=[]
for sentence in faqs.split("\n"):
  tokenized_dentence=tokenizer.texts_to_sequences([sentence])[0]
  print(tokenized_dentence)
  for i in range(1,len(tokenized_dentence)):
    n_gram=tokenized_dentence[:i+1]
    input_sequences.append(n_gram)
input_sequences
```

```python
max_len=max([len(x) for x in input_sequences])
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
input_sequences=pad_sequences(input_sequences,maxlen=max_len,padding="pre")
input_sequences
```

```python
x=input_sequences[:,:-1]
y=input_sequences[:,-1]
print(x.shape)
```

```python
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=len(tokenizer.word_index)+1)

print(y.shape)
print(y)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
```

```python
model=Sequential()
model.add(Embedding(len(tokenizer.word_index)+1,100,input_length=max_len-1))
model.add(LSTM(150))
model.add(Dense(len(tokenizer.word_index)+1,activation="softmax"))
```

```python
model.compile(loss="categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
model.summary()
```

```python
model.fit(x,y,epochs=100)
```

```python
text="machine learning and deep lerning"
#tokenize
for i in range(20):
      token_text=tokenizer.texts_to_sequences([text])[0]
#padd
      pad_text=pad_sequences([token_text],maxlen=max_len-1,padding="pre")
#prediction

      import numpy as np

      pos=np.argmax(model.predict(pad_text))
      for word,index in tokenizer.word_index.items():
          if index==pos:
             text=text+" "+word
             print(text)
```
